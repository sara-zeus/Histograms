```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
table=pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
table
```

```python
plt.hist(table["total_rooms"], bins=50, color="green" )
plt.hist(table["population"], bins=50, color="blue" )

#plt.show()
```

```python
import seaborn as sns
```

```python
sns.histplot(table["total_rooms"], bins=50, color="green" )

sns.histplot(table["population"], bins=50, color="blue" )


plt.show()
```

```python
sns.scatterplot(x= "total_rooms", y= "population", data=table)
```

```python
sns.kdeplot(table["total_rooms"], color="green",fill= True )

sns.kdeplot(table["population"], color="blue", fill=True )


plt.show()
```

```python
sns.kdeplot(data=table, x= "total_rooms", y= "population", fill=True)
```
